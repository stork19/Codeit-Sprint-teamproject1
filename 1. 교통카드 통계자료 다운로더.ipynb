```python
import os
import requests

# 다운로드 URL 템플릿
base_url = 'https://www.t-money.co.kr/ncs/pct/ugd/DownloadTrcrStstExl.dev?useYm={}'

# 시작 연월과 종료 연월 설정
start_year = 2014
start_month = 1
end_year = 2024
end_month = 6

# 다운로드할 연월 생성
def generate_year_month_range(start_year, start_month, end_year, end_month):
    year_month_list = []
    for year in range(start_year, end_year + 1):
        for month in range(1, 13):
            if (year == start_year and month < start_month) or (year == end_year and month > end_month):
                continue
            year_month_list.append(f"{year}{month:02d}")
    return year_month_list

# 파일 다운로드 함수
def download_file(year_month):
    download_url = base_url.format(year_month)
    file_response = requests.get(download_url)
    file_response.raise_for_status()
    
    # 년도별 폴더 생성
    year = year_month[:4]
    if not os.path.exists(year):
        os.makedirs(year)
    
    filename = os.path.join(year, f'{year_month}_교통카드_통계자료.xls')
    with open(filename, 'wb') as file:
        file.write(file_response.content)
    
    print(f"파일이 {filename} 이름으로 저장되었습니다.")

# 연월 리스트 생성
year_month_range = generate_year_month_range(start_year, start_month, end_year, end_month)

# 파일 다운로드
for year_month in year_month_range:
    try:
        download_file(year_month)
    except requests.HTTPError as e:
        print(f"파일 다운로드 실패: {year_month}, 오류: {e}")
```
