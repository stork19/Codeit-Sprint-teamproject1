```python
location_df = pd.read_csv('지하철역_좌표.csv', encoding='cp949')
```

```python
replace_dict = {
    '잠실' :  '잠실(송파구청)',
    '서울': '서울역',
    '삼성' : '삼성(무역센터)',
    '서울대입구' : '서울대입구(관악구청)'
}

location_df['역이름'] = location_df['역이름'].replace(replace_dict)
```

```python
# location_df[df_2024_01_sort[:20]['지하철역']]
df_2024_01_location = location_df[location_df['역이름'].isin(df_2024_01_sort[:20]['지하철역'])]
```

```python
df_2024_01_upper = df_2024_01_sort[:20]
```

```python
df_2024_01_upper['지하철역'].values
```

```python
# df_{년도}_{월}_upper(상위 20개 지역의 일평균)과 동일한 카테고리 순서로 맞출려고 지정
order = ['강남', '잠실(송파구청)', '홍대입구', '구로디지털단지', '신림', '서울역', '삼성(무역센터)', '역삼',
       '선릉', '신도림', '을지로입구', '고속터미널', '서울대입구(관악구청)', '영등포', '성수',
       '가산디지털단지', '용산', '사당', '수원', '종각']
```

```python
df_2024_01_location['역이름'] = pd.Categorical(df_2024_01_location['역이름'], categories=order, ordered=True)
df_2024_01_location = df_2024_01_location.sort_values('역이름')
```

```python
df_2024_01_location = df_2024_01_location.rename(columns={'역이름' : '지하철역'})
```

```python
df_2024_01_location.shape
```

```python
df_2024_01_upper.reset_index(drop=True, inplace=True)
df_2024_01_location.reset_index(drop=True, inplace=True)
```

```python
df_2024_01_upper[['y', 'x']] = df_2024_01_location[['y', 'x']]
```

```python
# y = 위도, x = 경도
df_2024_01_upper
```

```python
# 서울 중심부의 위도, 경도
center = [37.541, 126.986]

# 지도 생성
map = folium.Map(location=center, zoom_start=12)

# 지하철역별로 마커 추가
for idx, row in df_2024_01_upper.iterrows():
    # 이용객 수에 따라 마커의 크기를 다르게 설정
    radius = row['승하차승객수'] / 10000

    # 마커 생성
    marker = folium.CircleMarker(
        location=[row['y'], row['x']],  # 'location' 변수를 여기에 사용
        radius=radius,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f"{row['지하철역']}: 일 평균 이용객{row['승하차승객수']}명"
    )

    # 마커를 지도에 추가
    marker.add_to(map)

# 지도 저장
map.save('seoul_subway_usage.html')
map
```
